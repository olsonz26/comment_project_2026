### Comment Writing Engine

#### an application to help teachers write better comments in less time


```python
# functions and code for reading data files, generating comments and writing them to files
#Learning Indicators

from collections import Counter

data = []
class_data = {}

with open('teacher.csv') as file:
    header = file.readline().strip().split(',')
    #print(header)
    for line in file:
        parts = line.strip().split(',')
        data.append(parts)
        class_data[parts[0]] = class_data.get(parts[0], []) + parts[1:]

print(class_data)        
title = "\033[1mPython III: Data Science - Pigg A:\033[0m"
```

```python
indicators = ['identifying/characterizng/defining a programming problem', 
              'understanding and being able to use a variety of data types, control structures, and algorithmic problem-solving techniques in your programming', 
              'designing, documenting, implementing, and testing solutions to programming problems', 
              'expressing creativity using coding and technology in a variety of contexts', 
              'identifying and repairing coding errors in a program', 
              'effectively communicating programming solutions to others', 
              'working collaboratively to write modular code on larger projects']



def strength(student= "Zach"):
    """
    Takes in a student
    Returns which indicator is their strongest
    """
    indicator = class_data[student][-3]

    s_comment = f"You also always {class_data[student][-1]} in class."
    
    return f"A learning outcome at which you show strength is {indicators[int(indicator)-1]}", s_comment


print(strength('Taylor'))


def improve(student="Zoë"):
    """
    Takes in a student
    Returns which indicator is their weakest
    """
    indicator = class_data[student][-2]
    return f"A learning outcome at which you could improve in the future is working to {indicators[int(indicator)-1]}."


print(improve('Taylor'))
```

```python
def description():
    description = "This course builds on skills and concepts learned in our previous computer science courses. The topics have challenged you to explore how computing and technology can impact the world, with a unique focus on creative problem solving and real-world applications. As a student, you should know more advanced data structures and object-oriented programming basics. You also learned to communicate with popular APIs including Twitter and Google in order to “grab” data and to organize the data and display it."
    return description


description()
```

```python
def trend(student):
    """
    Calculates if the student had a positive or negative trejectory in the class.
    """
    big_improvement = "You've shown amazing growth with your exams over the course of this semester!"
    small_improvement = "You've shown some growth with your exams over the course of this semester."
    data = class_data[student]
    test1 = data[0]
    test2 = data[1]
    test3 = data[2]
    if test1 < test2 < test3:
        return big_improvement
    else:
        if test1 < test2 or test2 < test3:
            return small_improvement
        else:
            return None
print(trend('Taylor'))        

#'Taylor': ['77', '97', '84', '89', 'focused', '3', '4', 'good test results']
```

```python
def grade(student):
    """
    Writes a sentence about the student's final grade in the class.
    """
    data = class_data[student]
    final = int(data[3])
    if final >= 93 and final <= 100:
        letter = "A"
    elif final >= 90 and final < 93:
        letter = "A-"
    elif final >= 87 and final < 90:
        letter = "B+"   
    elif final >= 83 and final < 87:
        letter = "B" 
    elif final >= 80 and final < 83:
        letter = "B-"     
    elif final >= 77 and final < 80:
        letter = "C+" 
    elif final >= 73 and fina < 77:
        letter = "C" 
    elif final >= 70 and final < 73:
        letter = "C-"
    elif final >= 67 and final < 70:
        letter = "D+" 
    elif final >= 63 and final < 67:
        letter = "D" 
    elif final >= 60 and final < 63:
        letter = "D-"  
    elif final < 60:
        letter = "F"     
            
    return f"Your final grade in this class is {final}%, which ends up being the letter grade {letter}."


print(grade('Sam O.'))
```

```python
def cont(student):
    """
    Decides weather you should tell a student to continue with other coding classes.
    """
    data = class_data[student]
    final = int(data[3])
    if final >= 87:
        return "As you think about and plan for courses next year, I hope you consider some of the Computer Science electives like Algorithms and Machine Learning!"
    else:
        return None
    
    
print(cont('Taylor'))    
```

```python
def final_comment():
    """
    Writes a personalized comment to a separate file for each student’s name
    """
    for student in class_data:
        student_name = student
        strength_comment, s_comment = strength(student)
        improve_comment = improve(student)
        trend_comment = trend(student)
        grade_comment = grade(student)
        desc = description()
        last = cont(student)
        full_comment = (f"{title}\n\n{student_name},\n{desc} \n\n{strength_comment}. {s_comment} \n{improve_comment} \n\n{trend_comment} {grade_comment} {last}")
        
        filename = student_name.lower() + ".txt"        
        with open(filename, "w") as f:
            f.write(full_comment)
#         test
#         with open(filename) as c:
#             for i in c:
#                 print(i)
    print("All comments generated!")       
        
final_comment()        
```
